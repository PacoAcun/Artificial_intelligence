# 02. Preparacion de data

```python
import os
import cv2
import numpy as np
from tqdm import tqdm
```

```python

RAW_CHARACTERS = '../data/raw/character_dataset'
PROCESSED_OCR = '../data/processed/train_ocr'

TARGET_SIZE = (32, 32)
```

```python
def process_characters(input_dir, output_dir):
    if not os.path.exists(input_dir):
        print(f'Warning: Input directory {input_dir} does not exist.')
        return

    labels = [d for d in os.listdir(input_dir) if os.path.isdir(os.path.join(input_dir, d))]
    
    for label in tqdm(labels, desc='Processing characters'):
        label_input_dir = os.path.join(input_dir, label)
        label_output_dir = os.path.join(output_dir, label)
        os.makedirs(label_output_dir, exist_ok=True)
        
        files = os.listdir(label_input_dir)
        for filename in files:
            img_path = os.path.join(label_input_dir, filename)
            
            img = cv2.imread(img_path)
            if img is None:
                continue
            
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            
            resized = cv2.resize(gray, TARGET_SIZE)
            
            output_path = os.path.join(label_output_dir, filename)
            cv2.imwrite(output_path, resized)
```

```python
process_characters(RAW_CHARACTERS, PROCESSED_OCR)
print(f'Character processing complete. Data saved to {PROCESSED_OCR}')
```
